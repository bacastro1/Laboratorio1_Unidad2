```python
'''Se se procede a importar los modulos necesarios'''
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
'''Se cargan los datos del dataset'''
df = pd.read_csv('https://md-datasets-public-files-prod.s3.eu-west-1.amazonaws.com/e205d80e-2bc6-49ed-bfcc-4215b6b516fd')
df
```

```python
'''Devuelve valores únicos basados en una tabla hash en CLASS'''
df['CLASS'].unique()
```

```python
'''Se define si es Y o N con el uso del mapeo para cambiar a 1 si es Y y a 0 si es N.'''
d1 = {'Y': 1 , 'Y ': 1, 'N': 0, 'N ':0, 'P':2}
df['CLASS'] = df['CLASS'].map(d1)
df
```

```python
'''Devuelve valores únicos basados en una tabla hash en Gender'''
df['Gender'].unique()
```

```python
'''Se define en Gender si es F o M que significa masculino y femenino con el uso del mapeo para cambiar a 1 si es M y a 0 si es F.'''
d1 = {'F': 0 , 'f': 0, 'M': 1}
df['Gender'] = df['Gender'].map(d1)
df
```

```python
feature_cols = ['Gender', 'AGE', 'Urea', 'Cr', 'HbA1c', 'Chol', 'TG', 'HDL', 'LDL', 'VLDL', 'BMI']
X = df[feature_cols]
y = df['CLASS'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X_train, y_train)
y_pred = dtree.predict(X_test)
```

```python
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
print("Precision:", metrics.precision_score(y_test, y_pred, average='micro'))
print("Recall:", metrics.recall_score(y_test, y_pred, average='micro'))
```

```python
cm = confusion_matrix(y_test, y_pred, labels=dtree.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=dtree.classes_)
disp.plot()
plt.show()
```
